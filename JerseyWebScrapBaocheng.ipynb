```python
import os
import requests
from bs4 import BeautifulSoup as bs
from urllib.parse import urljoin

url = "https://baocheng3f888.x.yupoo.com/categories"
session = requests.Session()

# Create a folder to save the images
output_folder = "images"
os.makedirs(output_folder, exist_ok=True)

while True:
    reque = session.get(url)
    bss = bs(reque.content, "html.parser")

    categories_box = bss.find("div", class_="categories__box-right-main")
    if not categories_box:
        print("Categories box not found on the page.")
        break

    pg1 = categories_box.find_all(class_="categories__parent album__categories-box")
    if not pg1:
        print("Categories parent not found on the page.")
        break

    pgg1 = pg1[0].find_all(class_="categories__children")

    for i in pgg1:
        link = i.a.get('href', '')
        textplink = i.find_all("div", class_="text_overflow album__title")

        # Extract image URL
        img_tag = i.find("img", class_="card__thumb")
        if img_tag:
            image_data_src = img_tag.get('data-origin-src', '')
            if image_data_src:
                base_url = "https:"
                image_url = urljoin(base_url, image_data_src)
                print("Image URL:", image_url)

                # Download and save the image
                try:
                    image_response = session.get(image_url)
                    image_filename = os.path.join(output_folder, f"{os.path.basename(image_url)}")
                    with open(image_filename, "wb") as image_file:
                        image_file.write(image_response.content)
                    print(f"Image saved: {image_filename}")
                except Exception as e:
                    print(f"Error downloading image: {e}")

        if link and textplink:
            for title_div in textplink:
                if title_div.text:
                    category_url = f"https://baocheng3f888.x.yupoo.com{link}"
                    print(category_url)
                    reque_category = session.get(category_url)
                    bss_category = bs(reque_category.content, "html.parser").find(class_="showalbum__parent showalbum__nor nor").find_all(class_="showalbum__children image__main")

                    for i_category in bss_category:
                        divs_category = i_category.find_all("div")
                        for div_category in divs_category:
                            nested_div = div_category.find("div")
                            if nested_div and 'data-photoid' in nested_div.attrs:
                                data_photoid = nested_div['data-photoid']
                                emId = f"https://baocheng3f888.x.yupoo.com/{data_photoid}?uid=1"
                                print(emId)

                                # Download and save the image
                                try:
                                    image_response = session.get(emId)
                                    image_filename = os.path.join(output_folder, f"{data_photoid}.jpg")
                                    with open(image_filename, "wb") as image_file:
                                        image_file.write(image_response.content)
                                    print(f"Image saved: {image_filename}")
                                except Exception as e:
                                    print(f"Error downloading image: {e}")

    pg2 = bss.find(class_="none_select pagination__buttons")
    if pg2:
        next_button = pg2.find_all(class_="pagination__button")[-1]
        next_button_link = next_button.get('href', '')
        url = f"https://baocheng3f888.x.yupoo.com{next_button_link}"
    else:
        print("Pagination buttons not found on the page.")
        break
```
